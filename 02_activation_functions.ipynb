```
import numpy as np
import torch
import matplotlib.pyplot as plt
import torch.nn.functional as F
```

### Settings

```
def plot_function(x,y,label):
    plt.figure(figsize=(20, 5))
    plt.subplot(2, 3, 1)
    plt.plot(x, y, label=label)
    plt.title(label)
    plt.grid()
```

```
x = np.linspace(-10, 10, 100)
```

### Sigmoid

$$ \text{Sigmoid}(x) = \frac{1}{1 + e^{-x}} $$

```
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```
y_sigmoid = sigmoid(x)
plot_function(x,y_sigmoid,'Sigmoid')
```

### Tanh (Hyperbolic Tangent)

$$ \tanh(x) = \frac{e^{x} - e^{-x}}{e^{x} + e^{-x}} $$

```
def tanh(x):
    return np.tanh(x)
```

```
y_tanh = tanh(x)
plot_function(x,y_tanh,'Tanh')
```

### Softmax

$$ \text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}} $$

```
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)
```

```
y_softmax = softmax(x)
plot_function(x,y_softmax,'Softmax')
```

## Softplus

$$ \text{Softplus}(x) = \ln(1 + e^x) $$

```
def softplus(x):
    return np.log1p(np.exp(x))
```

```
y_softplus = softplus(x)
plot_function(x,y_softplus,'Softplus')
```

### ReLU

$$ \text{ReLU}(x) = \max(0, x) $$

```
def relu(x):
    return np.maximum(0, x)
```

```
y_relu = relu(x)
plot_function(x,y_relu,'ReLU')
```
